```python
pts = {(1,9), (1.25,7), (2,4), (4,2), (7,1), (6,1), (9, .75),
       (6,6), (4,5), (7,8), (9, 5), (3,7), (5,8), (7,3)}
```

```python
import matplotlib.pyplot as plt

def plot(pts):
    xs = [pt[0] for pt in pts]
    ys = [pt[1] for pt in pts]
    
    plt.scatter(xs, ys)
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.xlabel('Objective X')
    plt.ylabel('Objective Y')
    plt.grid()
    plt.show()
```

```python
plot(pts)
```

```python
def dominates(p1, p2):
    """ Return TRUE if p1 dominates p2 """
    return (p1[0]<p2[0] and p1[1]<=p2[1]) or (p1[0]<=p2[0] and p1[1]<p2[1]) 
```

```python
dominates((1,5), (3,4))
```

```python
def find_nds(pts):
    """Find the pareto optimal non-dominated subset of points """
    dom = set()
    for p in pts:
        bad = {q for q in pts if dominates(p,q)}
        dom = dom.union(bad)
    return pts - dom  # non-dominated points
```

```python
nds = find_nds(pts)
plot(nds)
```

```python
from functools import reduce

def reduce_nds(S, p):
    return S - {q for q in S if dominates(p,q)}

nds = reduce(reduce_nds, pts, pts)

plot(nds)
```
